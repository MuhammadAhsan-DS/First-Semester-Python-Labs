```python
# QUESTION NUMBER 4:

class Restaurant:
    def __init__(self, name, cuisine):
        self.restaurant_name = name
        self.cuisine_type = cuisine
        self.number_served = 0

    def describe_restaurant(self):
        print(f"{self.restaurant_name} serves {self.cuisine_type} food.")

    def open_restaurant(self):
        print(f"{self.restaurant_name} is now open!")

    def set_number_served(self, number):
        self.number_served = number

    def increment_number_served(self, number):
        self.number_served += number

restaurant = Restaurant("Food House", "Pakistani")
restaurant.describe_restaurant()
restaurant.open_restaurant()

print("Served:", restaurant.number_served)
restaurant.number_served = 25
print("Updated served:", restaurant.number_served)
restaurant.set_number_served(50)
print("Set served:", restaurant.number_served)
restaurant.increment_number_served(20)
print("After increment:", restaurant.number_served)
```

```python
# QUESTION NUMBER 5:

class IceCreamStand(Restaurant):
    def __init__(self, name, cuisine, flavours):
        super().__init__(name, cuisine)
        self.flavours = flavours

    def describe_restaurant(self):
        print(f"{self.restaurant_name} serves the following ice cream flavours:")
        for flavour in self.flavours:
            print(flavour)
ice_cream = IceCreamStand("Hico", "Ice Cream", ["Mango", "Strawberry", "Vanilla", "Chocolate"])
ice_cream.describe_restaurant()
```
