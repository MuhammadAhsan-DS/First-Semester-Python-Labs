```python
#Question Number 1:

def check_credit_eligibility(income, credit_score):
    if income <= 30000 and 600 <= credit_score <= 699:
        print("Review required")
    elif income <= 50000 and credit_score <= 700:
        print("Eligible")
    else:
        print("Not eligible")


check_credit_eligibility(40000, 800)  
check_credit_eligibility(30000, 611)  
```

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_value(self):
        return self.price * self.quantity


# Create objects
p1 = Product("accessories", 200, 3)
p2 = Product("laptop", 100000, 2)
p3 = Product("handfree", 1500, 5)

# Store in list
products = [p1, p2, p3]

# Loop through products
for item in products:
    print(f"Product: {item.name}")
    print(f"Total Value: {item.total_value()}")
    print("------")
```
