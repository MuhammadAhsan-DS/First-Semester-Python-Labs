```python
 #Q1: Fibonacci Series (First 10 Elements)
fibonacci = [0, 1]
for i in range(8):  # Already have 2 elements, need 8 more
    next_number = fibonacci[-1] + fibonacci[-2]
    fibonacci.append(next_number)

print(fibonacci)
```

```python
#Q2: Prime Numbers from 1 to 100
primes = []
for num in range(2, 101):  # 1 is not a prime number
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print(primes)
```

```python
#Q3: Movie Ticket Price Calculator
#Version 1 Loop executes five times:
for i in range(5):
    age = int(input("Enter your age: "))
    if age < 3:
        print("No ticket required for children under 3 years.")
    elif age <= 12:
        print("The movie theatre ticket costs $10.")
    else:
        print("The movie theatre ticket costs $15.")
#Version 2  Using an active variable:
active = True

while active:
    age = int(input("Enter your age: "))
    if age < 3:
        print("No ticket required for children under 3 years.")
    elif age <= 12:
        print("The movie theatre ticket costs $10.")
    else:
        print("The movie theatre ticket costs $15.")

    again = input("Would you like to enter another age? Enter 'no' to quit: ")
    if again.lower() == 'no':
        active = False

```
